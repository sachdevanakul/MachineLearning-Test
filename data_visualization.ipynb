```python
#importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset
df = pd.read_csv("Titanic.csv")
```

```python
#checking null values
df.isnull().sum()
```

```python
#plotting histogram to visualize age 
plt.figure(figsize=(8, 5))
sns.histplot(df['age'].dropna(), bins=30, kde=True, color='blue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#plotting countplot to check number of survivals with gender
plt.figure(figsize=(6,5))
sns.countplot(x='sex', hue='survived', data=df, palette='coolwarm')
plt.title("Survival Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["Not Survived", "Survived"])
plt.show()
```

```python
# Boxplot with hue to differentiate survivors
plt.figure(figsize=(8, 5))
sns.boxplot(x='class', y='fare', hue='survived', data=df, palette="Set2")
plt.title("Fare Distribution by Passenger Class and Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
#plotting heatmap for numericals datatype
corr_matrix = df.corr(numeric_only=True) 
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
