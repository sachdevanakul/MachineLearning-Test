```python
#loading libraries
import pandas as pd
```

```python
#loading dataset usimg pandas
df =pd.read_csv("employee_data.csv")
```

```python
#checking shape of dataset like number of rows and columns
df.shape
```

```python
#checking null values in the dataset
df.isnull().sum()
```

```python
#getting breif information about the dataset
df.info()
```

```python
#displaying first 5 rows in the dataset
df.head()
```

```python
#droping duplicates present in the dataset
df.drop_duplicates(inplace=True)
```

```python
#filling missing values in the department column with not known
#filling missing values in the Name column with not known
#filling missing values in the status column with active
df["Department"].fillna("Not known",inplace=True)
df["Name"].fillna("Not known",inplace=True)
df["Status"].fillna("Active",inplace=True)
```

```python
#checking to see whether the missing value is filled or not
df.head()
```

```python
#filling salary column with mean
mean=df["Salary"].mean
df["Salary"].fillna(mean, inplace=True)
```

```python
#dropping remaining null values
df.dropna(inplace=True)
```

```python
#all data is cleaned
df.info()
```
